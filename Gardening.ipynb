```python
#Inhaltsverzeichnis: 
print(f"\033[1mInhaltsverzeichnis\033[0m")
print(f"\033[1mKapitel: 1\033[0m Pflanzenübersicht")
print(f"\033[1mKapitel: 2\033[0m Filtern nach Jahreszeit, Pflegehinweis, Wuchshöhe")
print(f"\033[1mKapitel: 3\033[0m Benutzerabfrage - interaktiv")
print(f"\033[1mKapitel: 4\033[0m Top 10 Pflanzen mit den meisten Arten")
print(f"\033[1mKapitel: 5\033[0m Blume")
print(f"\033[1mKapitel: 6\033[0m Quiz mit Belohnung - interaktiv")
print()
print()
print()
print("\033[1;32mKapitel 1: Pflanzenübersicht\033[0m")
print()
#Tabelle
print(f"\n\033[1mPflanzenübersicht\033[0m")
print()

# Header der Tabelle (fett)
print(f"\033[1m{'Pflanze':<15}{'Artenanzahl':<15}{'Anpflanzzeit':<25}{'Wuchshöhe':<15}{'Standort':<30}{'Pflegehinweise'}\033[0m")
print("-" * 145)   # Linie zur Trennung der Kopfzeilen

# Blumen
print(f"\n\033[1mBlumen\033[0m")
print(f"{'Lavendel':<15}{'ca.25':<15}{'Frühling':<25}{'30-80cm':<15}{'sonnig, durchlässig':<30}{'trocken halten'}")
print(f"{'Sonnenblume':<15}{'ca.70':<15}{'Frühling':<25}{'100-300cm':<15}{'sonnig':<30}{'bewässert, windgeschützt'}")
print(f"{'Rose':<15}{'>100':<15}{'Herbst, Frühling':<25}{'50-200cm':<15}{'sonnig, halbschattig':<30}{'regelmäßiger Schnitt, Schädlinge kontrollieren'}")
print(f"{'Lilie':<15}{'>100':<15}{'Herbst':<25}{'80-120cm':<15}{'sonnig, halbschattig':<30}{'Boden leicht feucht halten, frostempfindlich'}")
print(f"{'Ringelblume':<15}{'ca.20':<15}{'Frühling':<25}{'30-50cm':<15}{'sonnig':<30}{'geringe Pflege, lockerer Boden'}")
print(f"{'Lavatera':<15}{'ca.25':<15}{'Herbst':<25}{'10-15cm':<15}{'halbschattig':<30}{'Boden feucht und humusreich halten'}")
print(f"{'Astern':<15}{'ca.180':<15}{'Frühjahr':<25}{'20-150cm':<15}{'sonnig, halbschattig':<30}{'regelmäßig gießen'}")
print("-" * 145) 

# Obst (fett)
print(f"\n\033[1mObst\033[0m")
print(f"{'Erdbeere':<15}{'ca.20':<15}{'Frühling':<25}{'10-30cm':<15}{'sonnig':<30}{'mulchen, regelmäßig gießen'}")
print(f"{'Johannisbeere':<15}{'ca.50':<15}{'Frühling':<25}{'100-150cm':<15}{'sonnig, halbschattig':<30}{'Windschutz, regelmäßig gießen'}")
print("-" * 145)

# Gemüse (fett)
print(f"\n\033[1mGemüse\033[0m")
print(f"{'Tomate':<15}{'ca.13':<15}{'Frühling':<25}{'100-200cm':<15}{'sonnig':<30}{'hoher Wasserbedarf, regelmäßig düngen'}")
print(f"{'Zucchini':<15}{'1':<15}{'Frühling':<25}{'60-100cm':<15}{'sonnig':<30}{'hoher Wasser- und Platzbedarf'}")
print(f"{'Karotte':<15}{'ca.6':<15}{'Frühling, Spätsommer':<25}{'20-40cm':<15}{'sonnig':<30}{'lockerer, sandiger Boden'}")
print(f"{'Rucola':<15}{'ca.10':<15}{'Herbst, Frühling':<25}{'15-30cm':<15}{'sonnig, halbschattig':<30}{'regelmäßig ernten und feucht halten'}")
print(f"{'Radieschen':<15}{'ca.3':<15}{'Herbst, Frühling':<25}{'10-30cm':<15}{'sonnig':<30}{'feuchter, durchlässiger Boden'}")
print("-" * 145) 

# Kräuter (fett)
print(f"\n\033[1mKräuter\033[0m")
print(f"{'Salbei':<15}{'ca.900':<15}{'Frühling':<25}{'30-60cm':<15}{'sonnig, trocken':<30}{'mäßig gießen'}")
print(f"{'Basilikum':<15}{'ca.60':<15}{'Frühling':<25}{'20-50cm':<15}{'sonnig':<30}{'regelmäßig ernten, hoher Wasserbedarf'}")
print(f"{'Thymian':<15}{'ca.300':<15}{'Frühling':<25}{'10-30cm':<15}{'sonnig, trocken':<30}{'trockener Boden'}")
print(f"{'Pfefferminze':<15}{'ca.25':<15}{'Frühling':<25}{'30-90cm':<15}{'halbschattig':<30}{'feucht halten'}")
print("-" * 145)


# Daten speichern
pflanzen_liste = [
    ("Lavendel", "ca.25", "", "Frühling", "30-80cm", "sonnig, durchlässig", "trocken halten"),
    ("Sonnenblume", "ca.70", "", "Frühling", "100-300cm", "sonnig", "bewässert, windgeschützt"),
    ("Rose", ">100", "", "Herbst", "50-200cm", "sonnig, halbschattig", "regelmäßiger Schnitt, Schädlinge kontrollieren"),
    ("Lilie", ">100", "", "Herbst", "80-120cm", "sonnig, halbschattig", "Boden leicht feucht halten, frostempfindlich"),
    ("Ringelblume", "ca.20", "", "Frühling", "30-50cm", "sonnig", "geringe Pflege, lockerer Boden"),
    ("Lavatera", "ca.25", "", "Herbst", "10-15cm", "halbschattig", "Boden feucht und humusreich halten"),
    ("Astern", "ca.180", "", "Frühjahr", "120-150cm", "sonnig, halbschattig", "regelmäßig gießen"),
    ("Erdbeere", "ca.20", "", "Frühling", "10-30cm", "sonnig", "mulchen, regelmäßig gießen"),
    ("Johannisbeere", "ca.50", "", "Frühling", "100-150cm", "sonnig, halbschattig", "Windschutz, regelmäßig gießen"),
    ("Tomate", "ca.13", "", "Frühling", "100-200cm", "sonnig", "hoher Wasserbedarf, regelmäßig düngen"),
    ("Zucchini", "1", "", "Frühling", "60-100cm", "sonnig", "hoher Wasser- und Platzbedarf"),
    ("Kartoffel", "ca.6", "", "Frühling, Spätsommer", "20-40cm", "sonnig", "lockerer, sandiger Boden"),
    ("Rucola", "ca.10", "", "Herbst, Frühling", "15-30cm", "sonnig, halbschattig", "regelmäßig ernten und feucht halten"),
    ("Radieschen", "ca.31", "", "Herbst, Frühling", "10-30cm", "sonnig", "feuchter, durchlässiger Boden"),
    ("Salbei", "tca.900", "", "Frühling", "30-60cm", "sonnig, trocken", "mäßig gießen"),
    ("Basilikum", "tca.6", "", "Frühling", "20-50cm", "sonnig", "regelmäßig ernten, hoher Wasserbedarf"),
    ("Thymian", "tca.300", "", "Frühling", "10-30cm", "sonnig, trocken", "trockener Boden"),
    ("Pfefferminze", "ca.25", "", "Frühling", "30-90cm", "halbschattig", "feucht halten")
]
print()
print()
print("\033[1;32mKapitel 2: Filtern nach Anpflanzzeit, Pflegehinweise und Wuchshöhe\033[0m")
print()
#Filtern
# Pflanzen herausfiltern, die im Frühling angepflanzt werden
print(f"\n\033[1mPflanzen, die im Frühling angepflanzt werden können:\n\033[0m")
print(f"\033[1m{'Pflanze':<15} {'Anpflanzzeit':<20}\033[0m")  # Tabellenüberschrift
print("-" * 35)  # Trennlinie

# Inhalt füllen
for pflanze in pflanzen_liste:
    if "Frühling" in pflanze[3]:
        print(f"{pflanze[0]:<15} {pflanze[3]:<20}")
print()
print()

#Plfanzen hersaufiltern, die regelmäßig gegossen werden müssen
print(f"\n\033[1mPflanzen, die regelmäßig gegossen werden müssen:\n\033[0m")
print(f"\033[1m{'Pflanze' :<15} {'Pflegehinweise' :<20}\033[0m")
print("-" * 35)  # Trennlinie

#Inhalt füllen
for pflanze in pflanzen_liste:
    if "regelmäßig gießen" in pflanze[6]:
        print(f"{pflanze[0]:<15} {pflanze[6]:<20}")
print()
print()
# Plfanzen herausfiltern, die höher als 10cm sind
print(f"\n\033[1mPflanzen, die höher als 10 cm sind:\n\033[0m")
print(f"{'Pflanze' :<15} {'Wuchshöhe' :<20}")
print("-" * 35)  # Trennlinie

for pflanze in pflanzen_liste:
    # Extrahiere die Wuchshöhe und entferne "cm"
    hoehe_str = pflanze[4].replace("cm", "")
    
    # Nehme nur den minimalen Wert der Höhe, z.B. "10" aus "10-30"
    min_hoehe = int(hoehe_str.split("-")[0]) if "-" in hoehe_str else int(hoehe_str)
  
# Filtere Pflanzen mit einer Höhe größer als 10 cm
    if min_hoehe > 10:
        print(f"{pflanze[0]:<15} {pflanze[4]:<20}")

def filter_pflanzen_nach_anpflanzzeit(pflanzen_liste, anpflanzzeit):
    gefilterte_pflanzen = []

    for pflanze in pflanzen_liste:
        # Überprüfen, ob der Index 3 existiert und String-Konvertierung verwenden
        if len(pflanze) > 3 and anpflanzzeit in str(pflanze[3]):
            gefilterte_pflanzen.append(pflanze)

    return gefilterte_pflanzen

# Beispiel-Daten
pflanzen_liste = [
    ("Lavendel", "ca.25", "", "Frühling", "30-80cm", "sonnig, durchlässig", "trocken halten"),
    ("Sonnenblume", "ca.70", "", "Frühling", "100-300cm", "sonnig", "bewässert, windgeschützt"),
    ("Rose", ">100", "", "Herbst", "50-200cm", "sonnig, halbschattig", "regelmäßiger Schnitt, Schädlinge kontrollieren"),
    ("Lilie", ">100", "", "Herbst", "80-120cm", "sonnig, halbschattig", "Boden leicht feucht halten, frostempfindlich"),
    ("Ringelblume", "ca.20", "", "Frühling", "30-50cm", "sonnig", "geringe Pflege, lockerer Boden"),
    ("Lavatera", "ca.25", "", "Herbst", "10-15cm", "halbschattig", "Boden feucht und humusreich halten"),
    ("Astern", "ca.180", "", "Frühjahr", "120-150cm", "sonnig, halbschattig", "regelmäßig gießen"),
    ("Erdbeere", "ca.20", "", "Frühling", "10-30cm", "sonnig", "mulchen, regelmäßig gießen"),
    ("Johannisbeere", "ca.50", "", "Frühling", "100-150cm", "sonnig, halbschattig", "Windschutz, regelmäßig gießen"),
    ("Tomate", "ca.13", "", "Frühling", "100-200cm", "sonnig", "hoher Wasserbedarf, regelmäßig düngen"),
    ("Zucchini", "1", "", "Frühling", "60-100cm", "sonnig", "hoher Wasser- und Platzbedarf"),
    ("Kartoffel", "ca.6", "", "Frühling, Spätsommer", "20-40cm", "sonnig", "lockerer, sandiger Boden"),
    ("Rucola", "ca.10", "", "Herbst, Frühling", "15-30cm", "sonnig, halbschattig", "regelmäßig ernten und feucht halten"),
    ("Radieschen", "ca.31", "", "Herbst, Frühling", "10-30cm", "sonnig", "feuchter, durchlässiger Boden"),
    ("Salbei", "tca.900", "", "Frühling", "30-60cm", "sonnig, trocken", "mäßig gießen"),
    ("Basilikum", "tca.6", "", "Frühling", "20-50cm", "sonnig", "regelmäßig ernten, hoher Wasserbedarf"),
    ("Thymian", "tca.300", "", "Frühling", "10-30cm", "sonnig, trocken", "trockener Boden"),
    ("Pfefferminze", "ca.25", "", "Frühling", "30-90cm", "halbschattig", "feucht halten")
]
print()
print()
print("\033[1;32mKapitel 3: Benutzerabfrage- interaktiv\033[0m")
print()
# Benutzerabfrage
anpflanzzeit = input("Nach welcher Anpflanzzeit möchten Sie filtern? (z.B. 'Frühling'):")

# Methode aufrufen und Ergebnisse anzeigen
gefilterte_pflanzen = filter_pflanzen_nach_anpflanzzeit(pflanzen_liste, anpflanzzeit)

print(f"\n\033[1mPflanzen, die im {anpflanzzeit} angepflanzt werden können:\n\033[0m")
print(f"\033[1m{'Pflanze' :<15} {'Anpflanzzeit' :<20}\033[0m")
print("-" * 35)  # Trennlinie

for pflanze in gefilterte_pflanzen:
    if anpflanzzeit in pflanze[3]:
        print(f"{pflanze[0]:<15} {pflanze[3]:<20}")

print()
print()

print("\033[1;32mKapitel 3: Top 10 Pflanzen mit den meisten Arten\033[0m")
print()

import matplotlib.pyplot as plt
import re

# Pflanzenliste
pflanzen_liste = [
    ("Lavendel", "ca.25", "", "Frühling", "30-80cm", "sonnig, durchlässig", "trocken halten"),
    ("Sonnenblume", "ca.70", "", "Frühling", "100-300cm", "sonnig", "bewässert, windgeschützt"),
    ("Rose", ">100", "", "Herbst", "50-200cm", "sonnig, halbschattig", "regelmäßiger Schnitt, Schädlinge kontrollieren"),
    ("Lilie", ">100", "", "Herbst", "80-120cm", "sonnig, halbschattig", "Boden leicht feucht halten, frostempfindlich"),
    ("Ringelblume", "ca.20", "", "Frühling", "30-50cm", "sonnig", "geringe Pflege, lockerer Boden"),
    ("Lavatera", "ca.25", "", "Herbst", "10-15cm", "halbschattig", "Boden feucht und humusreich halten"),
    ("Astern", "ca.180", "", "Frühjahr", "120-150cm", "sonnig, halbschattig", "regelmäßig gießen"),
    ("Erdbeere", "ca.20", "", "Frühling", "10-30cm", "sonnig", "mulchen, regelmäßig gießen"),
    ("Johannisbeere", "ca.50", "", "Frühling", "100-150cm", "sonnig, halbschattig", "Windschutz, regelmäßig gießen"),
    ("Tomate", "ca.13", "", "Frühling", "100-200cm", "sonnig", "hoher Wasserbedarf, regelmäßig düngen"),
    ("Zucchini", "1", "", "Frühling", "60-100cm", "sonnig", "hoher Wasser- und Platzbedarf"),
    ("Kartoffel", "ca.6", "", "Frühling, Spätsommer", "20-40cm", "sonnig", "lockerer, sandiger Boden"),
    ("Rucola", "ca.10", "", "Herbst, Frühling", "15-30cm", "sonnig, halbschattig", "regelmäßig ernten und feucht halten"),
    ("Radieschen", "ca.31", "", "Herbst, Frühling", "10-30cm", "sonnig", "feuchter, durchlässiger Boden"),
    ("Salbei", "tca.900", "", "Frühling", "30-60cm", "sonnig, trocken", "mäßig gießen"),
    ("Basilikum", "tca.6", "", "Frühling", "20-50cm", "sonnig", "regelmäßig ernten, hoher Wasserbedarf"),
    ("Thymian", "tca.300", "", "Frühling", "10-30cm", "sonnig, trocken", "trockener Boden"),
    ("Pfefferminze", "ca.25", "", "Frühling", "30-90cm", "halbschattig", "feucht halten")
]

# Funktion zum Umwandeln der Artenanzahl in eine numerische Zahl
def parse_artenanzahl(arten_str):
    # Entfernt "ca.", "tca." oder ">" und wandelt den Text in eine Zahl um- 
    arten_str = arten_str.replace("ca.", "").replace("tca.", "").replace(">", "")
    try:
        return int(arten_str)
    except ValueError:
        return 0  # Im Falle eines ungültigen Werts wie ">" (falls keine Zahl vorhanden ist)

# Extrahieren der Pflanze und ihrer Artenanzahl
pflanzen_arten = [(pflanze[0], parse_artenanzahl(pflanze[1])) for pflanze in pflanzen_liste]

# Sortieren der Pflanzen nach Artenanzahl (absteigend)
pflanzen_arten_sorted = sorted(pflanzen_arten, key=lambda x: x[1], reverse=True)

# Wählen der Top 10 Pflanzen aus
top_pflanzen = pflanzen_arten_sorted[:10]

# Pflanzen und deren Artenanzahl für das Diagramm
pflanzen_namen = [pflanze[0] for pflanze in top_pflanzen]
arten_anzahl = [pflanze[1] for pflanze in top_pflanzen]

# Erstellen des Balkendiagramms
plt.figure(figsize=(12, 6))
plt.bar(pflanzen_namen, arten_anzahl, color='skyblue')

# Diagrammdetails
plt.title("Top 10 Pflanzen mit den meisten Arten", fontsize=16)
plt.xlabel("Pflanzen", fontsize=14)
plt.ylabel("Anzahl der Arten", fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

# Werte über den Balken anzeigen
for i, wert in enumerate(arten_anzahl):
    plt.text(i, wert + 1, str(wert), ha='center', fontsize=12)

# Diagramm anzeigen
plt.tight_layout()
plt.show()


import matplotlib.pyplot as plt
import numpy as np

def draw_stem(ax, x, y, height):
    """Zeichnet den Stängel der Blume."""
    ax.plot([x, x], [y, y + height], color="green", linewidth=3)

def draw_leaf(ax, x, y, size, angle):
    """Zeichnet ein Blatt."""
    # Drehung des Blatts um den Winkel
    leaf_x = x + size * np.cos(angle)
    leaf_y = y + size * np.sin(angle)
    ax.add_patch(plt.Circle((leaf_x, leaf_y), size * 0.5, color="green", alpha=0.7))

def draw_flower(ax, x, y, petal_count, petal_size):
    """Zeichnet die Blüte der Blume mit Blütenblättern."""
    # Blütenblätter
    angles = np.linspace(0, 2 * np.pi, petal_count, endpoint=False)
    for angle in angles:
        petal_x = x + petal_size * np.cos(angle)
        petal_y = y + petal_size * np.sin(angle)
        ax.add_patch(plt.Circle((petal_x, petal_y), petal_size * 0.5, color="pink", alpha=0.8))
    
    # Blütenkern
    ax.add_patch(plt.Circle((x, y), petal_size * 0.4, color="yellow"))

def draw_simple_flower():
    """Zeichneung der Blume"""
    fig, ax = plt.subplots(figsize=(6, 10))
    ax.set_xlim(-3, 3)
    ax.set_ylim(0, 10)
    ax.axis("off")  # Achsen ausblenden

    # Blume zeichnen
    stem_x, stem_y, stem_height = 0, 0, 6
    draw_stem(ax, stem_x, stem_y, stem_height)
    draw_leaf(ax, stem_x, stem_y + 3, size=1, angle=-np.pi / 4)
    draw_flower(ax, stem_x, stem_y + stem_height, petal_count=8, petal_size=1)

    plt.show()

draw_simple_flower()
print()
print()

print("\033[1;32mKapitel 5: Quiz mit Belohnung- interaktiv\033[0m")
print()

from IPython.display import YouTubeVideo, display

def quiz():
    # Hinweis für den Benutzer
    print("Willkommen zum Quiz! 🚀")
    print("Bitte beantworte die folgenden Fragen, indem du mit *a, **b* oder *c* antwortest.")
    print("Viel Erfolg!\n")

    # Fragen und Antworten
    fragen = [
        {"frage": "Welche Jahreszeit eignet sich besonders für die Anpflanzung von Tomaten? (a) Frühling, (b) Winter, (c) Herbst", "antwort": "a"},
        {"frage": "Welche Pflanze benötigt besonders viel Wasser? (a) Kaktus, (b) Sonnenblume, (c) Lavendel", "antwort": "b"},
        {"frage": "Was ist die ideale Wuchshöhe für Ringelblumen? (a) 10-30cm, (b) 30-50cm, (c) 50-70cm", "antwort": "b"}
    ]

    richtige_antworten = 0


def quiz():
    # Hinweis für den Benutzer
    print("\033[1mWillkommen zum Quiz! 🚀\033[0m")
    print("Bitte beantworte die folgenden Fragen, indem du mit \033[1ma\033[0m, \033[1mb\033[0m oder \033[1mc\033[0m antwortest.")
    print("Viel Erfolg!\n")

    # Fragen und Antworten
    fragen = [
        {"frage": "Welche Jahreszeit eignet sich besonders für die Anpflanzung von Tomaten? (a) Frühling, (b) Winter, (c) Herbst", "antwort": "a"},
        {"frage": "Welche Pflanze benötigt besonders viel Wasser? (a) Kaktus, (b) Sonnenblume, (c) Lavendel", "antwort": "b"},
        {"frage": "Was ist die ideale Wuchshöhe für Ringelblumen? (a) 10-30cm, (b) 30-50cm, (c) 50-70cm", "antwort": "b"}
    ]

    richtige_antworten = 0

    # Quiz-Schleife
    for i, frage in enumerate(fragen):
        while True:
            print(f"\033[1mFrage {i+1}\033[0m: {frage['frage']}")
            user_antwort = input("Deine Antwort (a, b oder c): ").lower().strip()  # Antwort des Nutzers (case-insensitive)

            if user_antwort in ['a', 'b', 'c']:
                break
            else:
                print("\033[1;31mUngültige Eingabe! Bitte wähle a, b oder c.\033[0m")

        if user_antwort == frage["antwort"]:
            print("\033[1mRichtig! 🎉\033[0m")
            richtige_antworten += 1
        else:
            print("\033[1mLeider falsch. 😔\033[0m")

    # Ergebnis prüfen
    if richtige_antworten == len(fragen):
        print("\n\033[1mHerzlichen Glückwunsch! Du hast alle Fragen richtig beantwortet. 🎉🎉🎉\033[0m")
        print("\033[1;32mHier ist deine Belohnung:\033[0m")
        # Erfolgs-Video anzeigen
        youtube_id = "tTWAUADXAWM"  # Video-ID
        display(YouTubeVideo(youtube_id, width=800, height=450))
    else:
        print(f"\nDu hast {richtige_antworten} von {len(fragen)} Fragen richtig beantwortet.")
        print("\033[1;31mLeider hast du verloren. 😢\033[0m")
# Quiz starten
quiz()
```
